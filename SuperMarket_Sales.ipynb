```python
#import required libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#load the dataset
df = pd.read_csv("SuperMarket Analysis.csv")
```

```python
#display the first 5 rows of the dataset
df.head() 
```

```python
#Display the dataset information
df.info()
```

```python
#display statistical summary
df.describe()
```

```python
#2data cleaning
```

```python
#1.check for missing values 
df.isnull().sum()
#No missing values were found in the dataset

#2.check data types
df.info()

#3.Convert Data Column To Datetime Format
df['Date']=pd.to_datetime(df['Date'])

#4. verify Data Types After Cleaning
df.info()
```

```python
#Task3 Exploratory Data Analysis
```

```python
# basic statistics for numerical columns (mean,median,min,max)
df.describe() 
```

```python
#frequency of each branch(branch analysis)
df['Branch'].value_counts()

#Bar chart for branch frequency
df['Branch'].value_counts().plot(kind='bar')
plt.title('Branch frequency')
plt.xlabel('Branch')
plt.ylabel('count')
plt.show()
```

```python
#(customer type analysis)
#frequency of customer types
df['Customer type'].value_counts()

#Bar chart for customer type distribution
df['Customer type'].value_counts().plot(kind='bar')
plt.title('Customer type distribution')
plt.xlabel('Customer type')
plt.ylabel('count')
plt.show()
```

```python
##Frequency of gender (gender analysis)
df['Gender'].value_counts()

#Bar chart for Gender distribution
df['Gender'].value_counts().plot(kind='bar')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('count')
plt.show()
```

```python
#(payment methon analysis)
#frequency of payment method
df['Payment'].value_counts()

#Bar chart for payment method distributions 
df['Payment'].value_counts().plot(kind="bar")
plt.title('Payment method distribution')
plt.xlabel('Payment Method')
plt.ylabel('count')
plt.show()
```

```python
#(line plot:sales over time )
#total sales over time 
sales_over_time = df.groupby('Date')['Sales'].sum().sort_index()
#sort_index()----->A To ensure the correct data order for time-based analysis
sales_over_time.plot()
plt.title('Total sales over time ')
plt.xlabel('Date')
plt.ylabel('total Sales')
plt.show()
```

```python
#(line plot :Rating over time)
#AVG rating over time 
rating_over_time = df.groupby('Date')['Rating'].mean()

rating_over_time.plot()
plt.title('Avarage Rating Over Time')
plt.xlabel('Date')
plt.ylabel('Rating')
plt.show()
```

```python
#Scatter plot:Scales VS Rating
plt.scatter(df['Rating'], df['Sales'])
plt.title('Sales VS Rating')
plt.xlabel('Rating')
plt.ylabel('Sales')
plt.show()
```

```python
#correlation heatmap
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Box Plot:Gross income by product line
sns.boxplot(x='Product line',y='gross income',data=df)
plt.title('Gross Income By Product Line')
plt.xticks(rotation=45)
plt.show()
```

```python
#Task 4 :Sales & Revenue Metrics
## this function will give us the mean,min,max,quartiles
df['Sales'].describe()
```

```python
#Total Sales by product line
sales_by_product=df.groupby('Product line')['Sales'].sum()
sales_by_product

#Bar chart for sales by product line()
sales_by_product.plot(kind='bar')
plt.title('Total Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Sales')
plt.show()
```
